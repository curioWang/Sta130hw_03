````python
import pandas as pd 
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
penguins
````

````python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Remove rows with missing values in flipper_length_mm
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Create a function to calculate statistical measures for flipper_length_mm per species
def get_stats(data):
    mean = data.mean()
    median = data.median()
    min_val = data.min()
    max_val = data.max()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    std_dev = data.std()
    return mean, median, min_val, max_val, q1, q3, std_dev

# Plot separate histograms for each species
species_list = penguins['species'].unique()

for species in species_list:
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Calculate stats
    mean, median, min_val, max_val, q1, q3, std_dev = get_stats(species_data)
    std_dev_min = mean - 2 * std_dev
    std_dev_max = mean + 2 * std_dev

    # Create a figure for each species
    fig = go.Figure()

    # Add histogram (without normalization to use counts)
    fig.add_trace(go.Histogram(x=species_data, name=f'{species}', opacity=0.6))

    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), annotation_text=f'Mean', annotation_position="top left")
    fig.add_vline(x=median, line=dict(color='green', dash='dash'), annotation_text=f'Median', annotation_position="top right")

    # Add vertical lines for range, interquartile range, and 2 standard deviations range
    # Range (Min to Max)
    fig.add_vline(x=min_val, line=dict(color='red', dash='solid'), annotation_text=f'Range', annotation_position="top right")
    fig.add_vline(x=max_val, line=dict(color='red', dash='solid'))

    # Interquartile Range (Q1 to Q3)
    fig.add_vline(x=q1, line=dict(color='orange', dash='dot'), annotation_text=f'IQR', annotation_position="top right")
    fig.add_vline(x=q3, line=dict(color='orange', dash='dot'))

    # Two standard deviations range (±2 Std Dev)
    fig.add_vline(x=std_dev_min, line=dict(color='purple', dash='dashdot'), annotation_text=f'±2 Std Dev', annotation_position="top right")
    fig.add_vline(x=std_dev_max, line=dict(color='purple', dash='dashdot'))

    # Add rectangles for the same ranges as visual aid
    # Range (Min to Max)
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor='lightgray', opacity=0.2, line_width=0)
    
    # Interquartile Range (Q1 to Q3)
    fig.add_vrect(x0=q1, x1=q3, fillcolor='orange', opacity=0.2, line_width=0)

    # Two standard deviations range (±2 Std Dev)
    fig.add_vrect(x0=std_dev_min, x1=std_dev_max, fillcolor='lightblue', opacity=0.2, line_width=0)

    # Update layout with increased width
    fig.update_layout(title=f"Flipper Length Distribution for {species} with Statistical Markers",
                      xaxis_title="Flipper Length (mm)",
                      yaxis_title="Count",  # Kept Count as Y-axis label
                      width=900,  # Increased width for better label spacing
                      barmode='overlay')

    # Show the plot for the species
    fig.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load penguins dataset
penguins = sns.load_dataset('penguins')

# Filter out rows with missing 'flipper_length_mm'
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Function to plot KDE with statistical measures
def plot_kde_with_stats(ax, data, species_name):
    # Calculate statistics
    mean = data.mean()
    median = data.median()
    std = data.std()
    min_val = data.min()
    max_val = data.max()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    # Plot KDE
    sns.kdeplot(data, fill=True, ax=ax)
    
    # Mark mean and median
    ax.axvline(mean, color='red', linestyle='--', label='Mean')
    ax.axvline(median, color='blue', linestyle='-', label='Median')

    # Mark range
    ax.axvline(min_val, color='green', linestyle='--', label='Min')
    ax.axvline(max_val, color='green', linestyle='--', label='Max')

    # Mark interquartile range (IQR)
    ax.axvspan(q1, q3, color='yellow', alpha=0.3, label='IQR')

    # Mark two standard deviations range
    ax.axvspan(mean - 2*std, mean + 2*std, color='purple', alpha=0.2, label='2 Std Dev')

    # Set title and labels
    ax.set_title(f'{species_name} Flipper Length KDE')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')

# Create subplots for each species
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# List of species
species_list = penguins['species'].unique()

# Plot for each species
for i, species in enumerate(species_list):
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    plot_kde_with_stats(axes[i], species_data, species)

# Add a legend to the first subplot
axes[0].legend()

plt.tight_layout()
plt.show()

````

1. Box Plots
Box plots provide a compact summary of the distribution of numerical data through their quartiles.

Advantages:
Efficient Summary: Box plots give a concise, clear summary of the data distribution, including median, spread, quartiles, and potential outliers.
Outlier Detection: They make it easy to identify outliers in the data.
Comparison: Box plots are great for comparing distributions across different groups or categories.
Robust to Skewness: Since they rely on quartiles, box plots are not overly sensitive to skewed data.
Space Efficient: They use less space compared to other methods, making it easier to compare multiple groups side by side.

Disadvantages:
Limited Detail: Box plots don't show the specific shape of the distribution, such as multimodalities (multiple peaks) or subtle nuances in the data.
No Frequency Information: They don’t show the exact number of observations in each range or bin, which can be important in some analyses.
Less Intuitive: For people unfamiliar with statistics, box plots can be harder to interpret compared to more straightforward plots like histograms.


2. Histograms
Histograms display the frequency distribution of a dataset by dividing data into intervals (bins).

Advantages:
Clear Frequency Representation: Histograms show how many data points fall within each bin, making it easy to see the concentration of data.
Easy to Understand: Histograms are intuitive for most users to interpret, as they visually display the data frequency in specific intervals.
Distribution Shape: Histograms provide a good sense of the shape of the data distribution, including multimodal, skewed, or symmetric patterns.
Customizable Bin Size: You can adjust the bin size to get a more detailed or broader view of the data distribution.

Disadvantages:
Bin Size Sensitivity: The appearance of the histogram can vary greatly depending on the bin size, leading to potential misinterpretation if the bins are not chosen carefully.
Less Precise: The histogram can hide exact values, as data points are grouped into bins rather than displayed individually.
Not Ideal for Small Datasets: With small datasets, histograms may give misleading insights because of the limited number of observations in some bins.
Difficult to Compare: It can be harder to directly compare multiple histograms side by side compared to box plots.


3. Kernel Density Estimators (KDEs)
KDEs estimate the probability density function of a continuous variable, providing a smooth curve to visualize the distribution.

Advantages:
Smooth Representation: KDEs provide a smooth, continuous estimate of the distribution, making them useful for identifying the overall shape without the roughness of histograms.
Detailed Visualization: They reveal more detailed features, such as multimodality (multiple peaks), which may be missed in histograms or box plots.
No Bin Size Sensitivity: Unlike histograms, KDEs don’t rely on arbitrary bin sizes, which reduces potential misrepresentation of data based on bin choices.
Effective for Large Data: KDEs work well with large datasets, giving a detailed overview of the distribution.

Disadvantages:
Bandwidth Sensitivity: KDEs are highly sensitive to the bandwidth (smoothing parameter). Too small a bandwidth can make the KDE curve noisy, while too large a bandwidth may oversmooth the data and hide important features.
Not Ideal for Small Data: KDEs can produce misleading results with small datasets due to oversmoothing or undersmoothing.
Less Intuitive: Like box plots, KDEs can be harder for non-expert users to understand, especially when interpreting probability density.
No Exact Frequency: Unlike histograms, KDEs don’t show the exact number of data points at specific values or intervals.

#Right Skewness (Positive Skewness):
In a right-skewed distribution, the tail extends to the right, meaning there are a few extreme large values that pull the mean toward the right.
In such distributions, mean > median because the mean is more sensitive to outliers and is pulled in the direction of the extreme values.
 Left Skewness (Negative Skewness):
In a left-skewed distribution, the tail extends to the left, meaning there are a few extreme small values that pull the mean toward the left.
In these distributions, mean < median because the mean is dragged down by the small outliers, while the median is relatively unaffected by them.


````python
# Step 1: Import necessary libraries
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px

# Step 2: Generate Right-Skewed Data (Gamma Distribution)
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Step 3: Visualize Right-Skewed Distribution
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed Distribution")
fig1.show(renderer="png")

# Step 4: Calculate Mean and Median for Right-Skewed Data
mean1 = sample1.mean()
median1 = np.quantile(sample1, 0.5)
print(f"Mean (Right-Skewed): {mean1}")
print(f"Median (Right-Skewed): {median1}")

# Step 5: Generate Left-Skewed Data (Negated Gamma Distribution)
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Step 6: Visualize Left-Skewed Distribution
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed Distribution")
fig2.show(renderer="png")

# Step 7: Calculate Mean and Median for Left-Skewed Data
mean2 = sample2.mean()
median2 = np.quantile(sample2, 0.5)
print(f"Mean (Left-Skewed): {mean2}")
print(f"Median (Left-Skewed): {median2}")

# Explanation:
# In right-skewed distributions, Mean > Median because the tail pulls the mean to the right.
# In left-skewed distributions, Mean < Median because the tail pulls the mean to the left.
````

#the reason it happens
The mean is the average of all data points and is affected by extreme values (outliers). In skewed distributions, extreme values in the tail drag the mean towards the tail.
The median, on the other hand, is the middle value of the data, which is resistant to outliers. Therefore, it stays more central in the distribution and is less affected by extreme values compared to the mean.

````python
import pandas as pd
import plotly.express as px

# Step 1: Load the Baby Names dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Step 2: Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex']  # Add gender to names to differentiate identical names for boys and girls

# Step 3: Rank the names within each year based on the 'percent' column
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)

# Step 4: Sort the data by 'name' and 'year'
bn = bn.sort_values(['name', 'year'])

# Step 5: Calculate the percentage change from the previous year for each name
bn['percent change'] = bn['percent'].diff()

# Reset the percent change for new names (first occurrence of a name in the dataset)
new_name = [True] + list(bn['name'][:-1].values != bn['name'][1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Step 6: Filter to restrict to common names (percent > 0.001)
bn = bn[bn['percent'] > 0.001]

# Step 7: Sort the data by 'year'
bn = bn.sort_values('year')

# Step 8: Create the scatter plot using Plotly Express
fig = px.scatter(
    bn, 
    x="percent change", 
    y="rank", 
    animation_frame="year", 
    animation_group="name",
    size="percent", 
    color="sex", 
    hover_name="name", 
    size_max=50, 
    range_x=[-0.005, 0.005]  # Remove log_x and range_y
)

# Step 9: Reverse the y-axis so that rank 1 appears at the top
fig.update_yaxes(autorange='reversed')

# Step 10: Display the figure (use 'png' renderer for GitHub and MarkUs submissions)
fig.show(renderer="png")
````
